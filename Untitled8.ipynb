```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Generate Sample Data (Simulating an imported CSV/Excel file)
np.random.seed(42)
demo_data = {
    "Date": pd.date_range(start="2026-01-01", periods=100, freq="D"),
    "Product": ["Laptop", "Smartphone", "Headphones", "Tablet", "Smartwatch"] * 20,
    "Category": ["Electronics", "Electronics", "Accessories", "Electronics", "Accessories"] * 20,
    "Sales_Units": np.random.randint(5, 30, size=100),
    "Revenue": np.random.randint(1000, 8000, size=100)
}
df = pd.DataFrame(demo_data)
df["Date"] = pd.to_datetime(df["Date"])

# 2. Create Interactive Slicers/Filters using ipywidgets
category_slicer = widgets.SelectMultiple(
    options=list(df["Category"].unique()),
    value=list(df["Category"].unique()),
    description='Category:',
    disabled=False
)

product_slicer = widgets.SelectMultiple(
    options=list(df["Product"].unique()),
    value=list(df["Product"].unique()),
    description='Products:',
    disabled=False
)

# Output area to dynamically render the dashboard layout
output_dashboard = widgets.Output()

# 3. Dashboard Update Function
def update_dashboard(*args):
    with output_dashboard:
        clear_output(wait=True)

        # Filter dataframe based on selections
        selected_cats = list(category_slicer.value)
        selected_prods = list(product_slicer.value)

        df_filtered = df[df["Category"].isin(selected_cats) & df["Product"].isin(selected_prods)]

        if df_filtered.empty:
            print("No data available for the selected filters.")
            return

        # Calculate KPIs
        total_revenue = df_filtered["Revenue"].sum()
        total_units = df_filtered["Sales_Units"].sum()
        avg_order = df_filtered["Revenue"].mean()

        # Display KPIs
        print("=" * 60)
        print(f"💰 Total Revenue: ${total_revenue:,.2f}  |  📦 Units Sold: {total_units:,}  |  📊 Avg Order: ${avg_order:,.2f}")
        print("=" * 60 + "\n")

        # Create Visualizations
        fig, axes = plt.subplots(1, 2, figsize=(16, 5))

        # Chart 1: Revenue Trends Over Time
        trend_df = df_filtered.groupby("Date")["Revenue"].sum().reset_index()
        sns.lineplot(data=trend_df, x="Date", y="Revenue", ax=axes[0], marker="o", color="b", linewidth=2)
        axes[0].set_title("📈 Revenue Trends Over Time", fontsize=14, fontweight='bold')
        axes[0].tick_params(axis='x', rotation=45)
        axes[0].grid(True, linestyle="--", alpha=0.6)

        # Chart 2: Top Performing Products
        product_df = df_filtered.groupby("Product")["Revenue"].sum().reset_index().sort_values(by="Revenue", ascending=False)
        sns.barplot(data=product_df, x="Product", y="Revenue", ax=axes[1], palette="viridis", hue="Product", legend=False)
        axes[1].set_title("🏆 Top Performing Products", fontsize=14, fontweight='bold')
        axes[1].grid(True, axis='y', linestyle="--", alpha=0.6)

        plt.tight_layout()
        plt.show()

        # Display Data Table Preview
        print("\n📋 Filtered Dataset View (Top 5 rows):")
        display(df_filtered.head())

# Link updates to widget alterations
category_slicer.observe(update_dashboard, 'value')
product_slicer.observe(update_dashboard, 'value')

# 4. Display UI Layout Layout
print("🎯 Hold 'Ctrl' (or 'Cmd') to select multiple items in the filters below:")
filter_ui = widgets.HBox([category_slicer, product_slicer])
display(filter_ui, output_dashboard)

# Initialize dashboard on startup
update_dashboard()
```
